```python
import pandas as pd
import numpy as np
# libraries for making count matrix and similarity matrix
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pickle
# reading the data from the preprocessed .csv file
data = pd.read_csv(r'C:\Users\HP\OneDrive\Desktop\mini-project\modular\recommendation_system\data\final_data.csv')

```

```python
data=data.drop(['Unnamed: 0','Unnamed: 0.1','Unnamed: 0.1.1'], axis = 1)
```

```python
data.head()
```

```python
X=data['comb']
X
```

```python
# creating a count matrix
cv = CountVectorizer()
count_matrix = cv.fit_transform(X)
```

```python
# creating a similarity score matrix
sim = cosine_similarity(count_matrix)
```

```python
def rcmd(m):
    if m not in data['movie_title'].unique():
        return('Sorry! The movie you requested is not in our database. Please check the spelling or try with some other movies')
    else:
        i = data.loc[data['movie_title']==m].index[0]
        lst = list(enumerate(sim[i]))
        lst = sorted(lst, key = lambda x:x[1] ,reverse=True)
        lst = lst[1:11] # excluding first item since it is the requested movie itself
        l = []
        for i in range(len(lst)):
            a = lst[i][0]
            
            l.append(data['movie_title'][a])
        return l
```

```python
rcmd("avatar")
```

```python
# saving the similarity score matrix in a file for later use
np.save('similarity_matrix', sim)

# saving dataframe to csv for later use in main file
data.to_csv('data.csv',index=False)
```
